```python
biserial_vals = pd.DataFrame(columns = ["score","p-value"])

for var in df_raw.columns.drop("target"):

  biserial_vals.loc[var] = sp.stats.pointbiserialr(df_raw[var], df_raw['target'] == "M")

biserial_vals = biserial_vals.sort_values(by=['p-value'], ascending=True)

biserial_vals

```

```python
def choose_vars(df_corr, df_vars):

  chosen_vars = []

  for index, row in df_vars.iterrows():
    if row["p-value"] < 0.05:
      flag = True
      for value in chosen_vars:
        if df_corr[index][value] > 0.7:
          flag = False
          break
      if flag:
        chosen_vars.append(index)



  return chosen_vars

variables = choose_vars(corrs, biserial_vals)
variables
```

```python
# train/test split



df_raw["target"] = (df_raw["target"] == "M")*1

df_raw["target"]


```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler



#set target and predictor variables
X = df_raw[variables]
y = df_raw["target"]

#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)




#create a pipeline to standardize data and fit model
pipeline = Pipeline([('scaler', StandardScaler()), ('dtc', DecisionTreeClassifier(random_state = 100))])



#grid search for best model
grid_search = GridSearchCV(pipeline,
                           param_grid = {"dtc__max_depth": [4,5,6,7,8,9,10,11],
                                         "dtc__min_samples_split": [2,3,4,5,6,7,8,9],
                                         "dtc__max_features": [3,4,5,6,7,8,9]},
                           cv = 20, scoring = "f1")




grid_search.fit(X, y)

print(grid_search.best_params_)
print(grid_search.best_score_)

best_tree = grid_search.best_estimator_
best_tree.fit(X_train, y_train)
y_pred = best_tree.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))



```

```python
#Support vector Machine

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

#set target and predictor variables
X = df_raw[variables]
y = df_raw["target"]

#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


pipeline = Pipeline([('scaler', StandardScaler()), ("svc",SVC())])

grid_search = GridSearchCV(estimator = pipeline,
                           param_grid = {"svc__kernel": ["linear", "poly", "rbf", "sigmoid"],
                                         "svc__C": [0.1, 1, 10, 100],
                                         "svc__gamma": ["scale", "auto"]},
                           cv = 8, scoring = "f1")


grid_search.fit(X, y)

print(grid_search.best_params_)
print(grid_search.best_score_)

best_svm = grid_search.best_estimator_
best_svm.fit(X_train, y_train)
y_pred = best_svm.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))



```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

#set target and predictor variables
X = df_raw[variables]
y = df_raw["target"]

#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


pipeline = Pipeline([('scaler', StandardScaler()), ("log",LogisticRegression())])

grid_search = GridSearchCV(estimator = pipeline,
                           param_grid = {"log__penalty": ["l1", "l2", "elasticnet", None],
                                         "log__solver": ["lbfgs", "liblinear", "newton-cg", "newton-cholesky", "sag", "saga"],
                                         "log__C" : [100, 10, 1.0, 0.1, 0.01]},
                           cv = 8, scoring = "f1")

grid_search.fit(X, y)

print(grid_search.best_params_)
print(grid_search.best_score_)

best_log = grid_search.best_estimator_
best_log.fit(X_train, y_train)
y_pred = best_log.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```

```python
#K Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

X = df_raw[variables]
y = df_raw["target"]


# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Standardizing the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# KNN classifier
knn = KNeighborsClassifier()

# Setting up the pipeline
pipeline = Pipeline([('scaler', StandardScaler()), ('knn', knn)])

# Grid search parameters
param_grid = {"knn__n_neighbors": list(range(1, 21,2)), # Trying n_neighbors from 1 to 20
              "knn__weights": ["uniform", "distance"], # Trying different weighting methods
              "knn__metric": ["euclidean", "manhattan", "minkowski"]} # Trying different distance metrics

# Performing grid search with cross-validation
grid_search = GridSearchCV(estimator=pipeline,
                           param_grid=param_grid,
                           cv=8, scoring="f1")

# Fitting the grid search
grid_search.fit(X, y)

# Best parameters and best score from the grid search
print(grid_search.best_params_)
print(grid_search.best_score_)

# Using the best found parameters to fit on training data
best_knn = grid_search.best_estimator_
best_knn.fit(X_train, y_train)
y_pred = best_knn.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))

```

```python
#Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import pandas as pd

# Assuming df_raw is your dataframe
X = df_raw[variables]
y = df_raw["target"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize the RandomForestClassifier with default parameters
rf = RandomForestClassifier(random_state=42)

# Define the parameter grid
param_grid = {
    'n_estimators': [10,20,30,40],
    'max_depth': [4, 5, 6, 7]
}

# Initialize GridSearchCV with the RandomForestClassifier and parameter grid
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=8, scoring='f1', n_jobs=-1)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters found: ", grid_search.best_params_)
print("Best F1 score: ", grid_search.best_score_)

# Evaluate the model on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Calculate and print metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")

```
